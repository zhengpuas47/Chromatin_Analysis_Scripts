```python
rid_2_th = {_id:0. for _id in ctp13_codebook['id']}
rid_2_matched_region = {_id:"" for _id in ctp13_codebook['id']}
for _fl in liftover_bed_files:
    _threshold = int(re.match(r'^ctp13_(?P<threshold>[0-9]+)\.bed$', os.path.basename(_fl), ).groupdict()['threshold']) / 10
    _threshold_df = pd.read_csv(_fl, delimiter='\t', header=None, names=['Chr','Start','End','Name'])
    for _i,_row in _threshold_df.iterrows():
        _rid = int(_row['Name'].split('Reg_')[1])
        #rid_2_th[_rid] = max(_threshold, rid_2_th[_rid])
        if _threshold > rid_2_th[_rid]:
            rid_2_th[_rid] = _threshold
            rid_2_matched_region[_rid] = f"{_row['Chr']}:{_row['Start']}-{_row['End']}"
```

```python
# convert into vector and append to codebook
threshold_vector = [rid_2_th[_rid] for _rid in ctp13_codebook['id']]
matched_region_vector = [rid_2_matched_region[_rid] for _rid in ctp13_codebook['id']]
ctp13_codebook['liftover_th'] = threshold_vector
ctp13_codebook['liftover_region'] = matched_region_vector
```
