```python
lib_readout_df = pd.read_excel(os.path.join(library_folder, 'Readout_Usage_raw.xlsx'))
lib_readout_df
```

```python
kwd_2_dtype = {_v:_k for _k,_v in ia.classes._allowed_kwds.items()}
```

```python
dtype_2_ReadoutNameDict = {}
for _info, _stv in zip(lib_readout_df['Readout_name'], lib_readout_df['Stv_name']):
    if _info[0] in kwd_2_dtype:
        if kwd_2_dtype[_info[0]] not in dtype_2_ReadoutNameDict:
            dtype_2_ReadoutNameDict[kwd_2_dtype[_info[0]]] = {}
        # append
        dtype_2_ReadoutNameDict[kwd_2_dtype[_info[0]]][int(_info[1:])] = _stv
for _info, _stv in zip(lib_readout_df['Readout_name.1'], lib_readout_df['Stv_name.1']):
    if not isinstance(_info,str):
        continue
    if _info[0] in kwd_2_dtype:
        if kwd_2_dtype[_info[0]] not in dtype_2_ReadoutNameDict:
            dtype_2_ReadoutNameDict[kwd_2_dtype[_info[0]]] = {}
        # append
        dtype_2_ReadoutNameDict[kwd_2_dtype[_info[0]]][int(_info[1:])] = _stv
```

```python
data_type = 'combo'
name_2_type, name_2_id = {}, {}

for _data_type in dtype_2_ReadoutNameDict:
    name_2_id.update({_v:_k for _k,_v in dtype_2_ReadoutNameDict[_data_type].items()})
    name_2_type.update({_v:ia.classes._allowed_kwds[_data_type] for _k,_v in dtype_2_ReadoutNameDict[_data_type].items()})
```

```python
readout_usage_info = pd.read_excel(os.path.join(library_folder, 'SI14_info.xlsx'))
```

```python
list(dtype_2_ReadoutNameDict['combo'].values())
```

```python
codebook = pd.DataFrame(columns=['name', 'id', 'chr', 'chr_order']+list(dtype_2_ReadoutNameDict['combo'].values()))
```

```python
codebook['id'] = readout_usage_info.index
codebook['name'] = [_c.split('chr')[1] for _c in readout_usage_info['Coordinates']]
for _i, _stvs in zip(readout_usage_info.index, readout_usage_info['STV_names']):
    _stvs = _stvs.split('--')
    codebook.loc[codebook['id']==_i, _stvs] = 1

# chr info
for _i, _chr_info in zip(readout_usage_info.index, readout_usage_info['Name']):
    _chr = _chr_info.split("__")[0].split('chr')[1]
    _chr_ord = int(_chr_info.split("__")[1]) - 1
    codebook.loc[codebook['id']==_i, ['chr', 'chr_order']] = [_chr, _chr_ord]
    
codebook = codebook.fillna(0)
codebook.sort_values('id', inplace=True)
codebook.astype({_r:int for _r in dtype_2_ReadoutNameDict['combo'].values()})
```

```python
# Check usage per bit or per region
codebook_matrix = codebook[list(dtype_2_ReadoutNameDict['combo'].values())].values
np.unique(codebook_matrix.sum(0)), np.unique(codebook_matrix.sum(1))
```

```python
codebook_filename = os.path.join(library_folder, 'SI14_codebook.csv')
print(codebook_filename)
codebook.to_csv(codebook_filename, index=None)
```
