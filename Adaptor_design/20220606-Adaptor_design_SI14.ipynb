# Read SI14 information

```python
import pandas as pd
```

```python
library_folder = r'X:\Chromatin_Libraries\DNA-MERFISH_Seon\SI14'
```

```python
lib_readout_df = pd.read_excel(os.path.join(library_folder, 'Readout_Usage_raw.xlsx'))
lib_readout_df
```

```python
kwd_2_dtype = {_v:_k for _k,_v in ia.classes._allowed_kwds.items()}
```

```python
dtype_2_ReadoutNameDict = {}
for _info, _stv in zip(lib_readout_df['Readout_name'], lib_readout_df['Stv_name']):
    if _info[0] in kwd_2_dtype:
        if kwd_2_dtype[_info[0]] not in dtype_2_ReadoutNameDict:
            dtype_2_ReadoutNameDict[kwd_2_dtype[_info[0]]] = {}
        # append
        dtype_2_ReadoutNameDict[kwd_2_dtype[_info[0]]][int(_info[1:])] = _stv
for _info, _stv in zip(lib_readout_df['Readout_name.1'], lib_readout_df['Stv_name.1']):
    if not isinstance(_info,str):
        continue
    if _info[0] in kwd_2_dtype:
        if kwd_2_dtype[_info[0]] not in dtype_2_ReadoutNameDict:
            dtype_2_ReadoutNameDict[kwd_2_dtype[_info[0]]] = {}
        # append
        dtype_2_ReadoutNameDict[kwd_2_dtype[_info[0]]][int(_info[1:])] = _stv
```

# Load all Stvs

```python
all_stv_df = pd.read_excel(os.path.join(library_folder, 'allStvs.xlsx'))
all_stv_df
```

# Generate dtype_2_readoutseqdict

```python
dtype_2_ReadoutRecords = {}
for _dtype, _name_dict in dtype_2_ReadoutNameDict.items():
    _records = []
    for _rid, _name in sorted(_name_dict.items()):
        _seq = all_stv_df.loc[all_stv_df['Name']==_name, 'Sequence'].values[0]
        _record = SeqRecord(_seq, id=_name, name=_name, description='')
        _records.append(_record)
        
    dtype_2_ReadoutRecords[_dtype] = _records
```

```python
dtype_2_ReadoutRecords['combo']
```

## load readout sites

```python
# load readout sites
master_folder = r'\\10.245.74.212\Chromatin_NAS_2\Chromatin_Libraries'
adaptor_folder = os.path.join(master_folder, 'Adaptors')
readout_site_file = os.path.join(adaptor_folder, 'Readout_sites.fasta')
readout_sites = []
readout_site_dict = {}
readout_channels = ['750', '647', '561']

with open(readout_site_file, 'r') as _rd_handle:
    for _i, _readout in enumerate(SeqIO.parse(_rd_handle, "fasta")):
        readout_sites.append(_readout)
        readout_site_dict[readout_channels[_i]] = _readout

print(readout_site_dict)
```

## generate adaptors

```python
from ImageAnalysis3.library_tools.readouts import Generate_adaptors
```

```python
adaptors = []

for _dtype, _records in dtype_2_ReadoutRecords.items():
    print(_dtype, len(_records))
    
    _adaptors = Generate_adaptors(_records, [readout_site_dict['750'],
                                             readout_site_dict['647'],
                                             readout_site_dict['561']])
    for _i in range(len(_adaptors)):
        _adaptors[_i].id = f"SI14{ia.classes._allowed_kwds[_dtype]}{_i+1}_" + _adaptors[_i].id
    
    adaptors += _adaptors
```

# Save

```python
save_df = pd.DataFrame([], columns=['Name', 'Sequence', 'Scale', 'Purification'])
for _adt in adaptors:
    _info = {'Name':_adt.id, 
             'Sequence':str(_adt.seq), 
             'Scale':'25nm', 
             'Purification':'STD'}
    save_df = save_df.append(_info, ignore_index=True)
```

```python
save_df.to_csv(os.path.join(adaptor_folder, 'Order_20220606_SI14_adaptors.csv'),
               index=None)
```
